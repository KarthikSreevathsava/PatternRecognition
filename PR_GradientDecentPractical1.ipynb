```python
import matplotlib.pyplot as plt
import numpy as np

# Cost function: J(theta) = theta^2
def cost_function(theta):
    return theta**2

# Gradient of the cost function: dJ(theta)/dtheta = 2*theta
def gradient(theta):
    return 2 * theta

# Gradient Descent Parameters
theta = 10  # Initial value
learning_rate = 0.2
iterations = 25

# Arrays to store theta and cost function values
theta_values = [theta]
cost_values = [cost_function(theta)]

# Gradient Descent Algorithm
for _ in range(iterations):
    theta -= learning_rate * gradient(theta)
    theta_values.append(theta)
    cost_values.append(cost_function(theta))

    print(f"Iteration: {len(cost_values) - 1}, Theta: {theta:.4f}, Cost: {cost_values[-1]:.4f}")
# Plotting the cost function
theta_range = np.linspace(-10, 10, 400)
cost_range = cost_function(theta_range)

plt.figure(figsize=(10, 6))
plt.plot(theta_range, cost_range, label='Cost Function: $J(\\theta) = \\theta^2$')
plt.scatter(theta_values, cost_values, color='red', label='Gradient Descent Steps')
plt.plot(theta_values, cost_values, color='red', linestyle='--')
plt.xlabel('$\\theta$')
plt.ylabel('$J(\\theta)$')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```
